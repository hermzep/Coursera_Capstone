```python
#converted to pandas
dfs = pd.read_html(listing)
data=dfs[0]
data.head()
data.shape
```

```python
#cleaning the data to prepare it to process
#if borough not assigned then it is taken off
data1=data[data['Borough']!='Not assigned']
data1.head()
```

```python
data1.columns
```

```python
#joining any repeat postal codes
data2 = data1.groupby(['Postal Code','Borough'], sort=False).agg(', '.join)
data2.reset_index(inplace=True)

# Replacing the name of the neighborhoods that are 'not assigned with Borough names
data2['Neighbourhood'] = np.where(data2['Neighbourhood'] == 'Not assigned',data2['Borough'], data2['Neighbourhood'])

data2
```

```python
data2.shape
```

```python
latnlong = pd.read_csv('https://cocl.us/Geospatial_data')
latnlong.head()
```

```python
data3 = pd.merge(data2,latnlong,on= 'Postal Code')
data3.head()
```

```python
data3
```

```python
!pip install folium
import folium
```

```python
#creating map of Toronto
maptoronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)
for lat,lng,borough,neighbourhood in zip(data3['Latitude'],data3['Longitude'],data3['Borough'],data3['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(maptoronto)
maptoronto
```

```python
#K clustering of neighborhoods

k=5
torontocluster = data3.drop(['Postal Code','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(torontocluster)
kmeans.labels_
data3.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
data3
```

```python
# create map
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(data3['Latitude'], data3['Longitude'], data3['Neighbourhood'], data3['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
